```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("plantCLFcollection_ObservationID.csv")
df
```

```python
df.CollectionId.unique()
```

## Mapping CollectionID

```python
collection = np.sort(df.CollectionId.unique())
collection_dict = {}
for i in range(1,df.CollectionId.nunique()+1):
    collection_dict[collection[i-1]] = i
```

```python
df["CollectionId_mapped"] = df.CollectionId.map(collection_dict)
df
```

```python
df[df.CollectionId_mapped == 1]
```

```python
df.CollectionId_mapped.nunique()
```

```python
df = df.drop(columns=["CollectionId"])
df
```

```python
df.rename(columns={"CollectionId_mapped": "CollectionId"},inplace=True)
df
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

df_Flower = df[df.Content == "Flower"]
df_Leaf = df[~df.index.isin(df_Flower.index)]

X, y = df_Flower.CollectionId, df_Flower.Genus

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=12)
```

```python
df_train = df_Flower[df_Flower.CollectionId.isin(X_train.values)].append(df_Leaf[df_Leaf.CollectionId.isin(X_train.values)]).reset_index(drop=True)
```

```python
df_test = df_Flower[df_Flower.CollectionId.isin(X_test.values)].append(df_Leaf[df_Leaf.CollectionId.isin(X_test.values)]).reset_index(drop=True)
```

```python
df_train.to_csv("train_data.csv",index=False)
df_test.to_csv("test_data.csv",index=False)
```
